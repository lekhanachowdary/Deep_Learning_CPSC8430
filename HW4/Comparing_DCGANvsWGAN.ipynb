```python
import numpy as np
import matplotlib.pyplot as plt

dcgan_fid = np.load('Results/FID_score/DC_FID.npy')
wgan_fid = np.load('Results/FID_score/WGAN_FID.npy')

epochs_dcgan = np.arange(len(dcgan_fid))
epochs_wgan = np.arange(len(wgan_fid))

#Plot the FID values
plt.figure(figsize=(10, 6))
plt.plot(epochs_dcgan, dcgan_fid, label='DCGAN', color='purple', marker='o')
plt.plot(epochs_wgan, wgan_fid, label='WGAN', color='orange', marker='x')
plt.title('FID Score Comparison: DCGAN vs WGAN')
plt.xlabel('Epochs')
plt.ylabel('FID Score')
plt.legend()
plt.grid(True)

plt.show()
```
